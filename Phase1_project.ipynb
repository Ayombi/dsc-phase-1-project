```python
#Studio
#proportion of the missing data >5%
sns.histplot(final_df['studio'])
plt.title('distribution curve for studio before cleaning')
```

```python
#Year
#proportion of the missing data >5%
sns.histplot(final_df['year'], kde = True)
plt.title('distribution curve for year before cleaning')
# The distribution is positively skewed
```

```python
#runtime_minutes
#proportion of the missing data >5%
plt.figure(figsize=(5, 10))
sns.histplot(final_df['runtime_minutes'], kde = True)
plt.title('distribution curve for runtime_minutes  before cleaning')
# The distribution is positively skewed
```

```python
#Unnecessary data
#We note that start_year is the same as year, therefore, we shall drop the column 'year'
final_df.drop('year',axis =1, inplace=True)
final_df.head()
```

```python
final_df.info()
```

```python
#genres and studio are categorical data and therefore, we shall fill in the missing values with a string 'Nan' to show the values are missing
final_df['genres'].fillna('Nan', inplace=True)
final_df['studio'].fillna('Nan', inplace=True)
```

```python
final_df.info()
```

```python
#domestic_gross
# We'll impute using the median because of the positive skewness of the distribution and the many missing  values.We will however not drop the column as domestic gross is key in the box office performance
# using the fillna function
final_df['domestic_gross'].fillna(final_df['domestic_gross'].median(), inplace =True)
```

```python
#Despite the fact the foreign gross is a key indficator in the movie performance, we will drop the column
final_df.drop('foreign_gross', axis=1, inplace=True)
```

```python
final_df.info()
```

```python
#runtime_minutes
#There appears to be some outliers due to the histogram plot. Lets plot a box plot
sns.boxplot(final_df['runtime_minutes'])
plt.title("Box plot of runtime_minutes")
```

```python
# drop outside the bounds
# Q1 = 25%
# Q2 = 50%
# Q3 = 75%
# Q4 = 100%
# IQR = Q3 -Q1
Q1 = final_df['runtime_minutes'].quantile(0.25)
Q3 = final_df['runtime_minutes'].quantile(0.75)
IQR = Q3 - Q1

#Creating the upper and lower bounds
upper_bound = Q3 + 1.5*IQR
lower_bound = Q1 - 1.5*IQR
```

```python
# drop any outliers
#using .clip function

final_df['runtime_minutes'] = final_df['runtime_minutes'].clip(lower = lower_bound, upper = upper_bound)
```

```python
# plotting after dropping outliers
sns.boxplot(final_df['runtime_minutes'])
plt.title('Plot of runtime_minutes after dropping the Outliers')
```

```python
final_df.info()
```

```python
sns.histplot(final_df['runtime_minutes'], kde = True)
plt.title('distribution curve for runtime_minutes  after removing outliers')
```

```python
#runtime_minutes distribution is now symmetricakl therefore we shall impute the missing values using the mean
final_df['runtime_minutes'].fillna(final_df['runtime_minutes'].mean(), inplace =True)
```
